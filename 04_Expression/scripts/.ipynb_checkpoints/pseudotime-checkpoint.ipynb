## Plot Trees

I wrote a custom function to plot the generated tree embeddings using `ggplot2`.

```r
plot.tree <- function(ddr.tree, color.factor=c(), factor.name="Factor") {
  
  Z = t(ddr.tree$Z) %>%
    as.data.frame() %>%
    dplyr::select(Comp1=1, Comp2=2)
  
  Y = t(ddr.tree$Y) %>%
    as.data.frame() %>%
    dplyr::select(Comp1=1, Comp2=2)
  
  tree = as.data.frame(summary(ddr.tree$stree)) %>%
    dplyr::filter(i < j) %>%
    dplyr::mutate(StartComp1=Y$Comp1[i], StartComp2=Y$Comp2[i], EndComp1=Y$Comp1[j], EndComp2=Y$Comp2[j])
  
  Y$Degree = table(c(tree$i, tree$j))[1:nrow(Y)]
  
  Y.branch.terminal.points = Y %>%
    dplyr::filter(Degree == 1)
    
  Y.branch.points = Y %>%
    dplyr::filter(Degree > 2)
  
  if (length(color.factor) == 0) {
    
    p = ggplot() +
      geom_point(data=Z, aes(x=Comp1, y=Comp2), alpha=0.25, color="royalblue4") + 
      geom_segment(data=tree, aes(x=StartComp1, y=StartComp2, xend=EndComp1, yend=EndComp2), size=I(2)) +
      geom_point(data=Y.branch.terminal.points, aes(x=Comp1, y=Comp2), size=I(4), color="firebrick3") +
      geom_point(data=Y.branch.points, aes(x=Comp1, y=Comp2), size=I(4), color="goldenrod2")
  } else if (class(color.factor) == "factor") {
    
    Z = Z %>%
      dplyr::mutate(Factor = color.factor)
    
    p = ggplot() +
      geom_segment(data=tree, aes(x=StartComp1, y=StartComp2, xend=EndComp1, yend=EndComp2), size=I(2), alpha=0.5) +
      geom_point(data=Z, aes(x=Comp1, y=Comp2, color=Factor)) +
      scale_color_manual(values=colorRampPalette(brewer.pal(8, "Pastel2"))(length(levels(Z$Factor)))) +
      guides(color=guide_legend(factor.name))
  } else {
    
    Z = Z %>%
      dplyr::mutate(Factor = color.factor)
    
    factor.max = max(abs(color.factor))
    
    p = ggplot() +
      geom_segment(data=tree, aes(x=StartComp1, y=StartComp2, xend=EndComp1, yend=EndComp2), size=I(2), alpha=0.5) +
      geom_point(data=Z, aes(x=Comp1, y=Comp2, color=Factor))
    
    if (all(color.factor >= 0)) {
      p = p + scale_color_viridis(name=factor.name, limits=c(0, factor.max), option="B")
    } else {
      p = p + scale_color_viridis(name=factor.name, limits=c(-factor.max, factor.max), option="C")
    }
  }
  
  p = p + 
    xlab("Component 1") + ylab("Component 2") +
    ggplot_theme +
    theme(legend.position="right")
  
  return(p)
}
```

The tree for the entire gene expression matrix. Each dot represents a branch point (a vertex with degree > 2). The black path represents the underlying graph.

```r
options(repr.plot.width=6, repr.plot.height=6)
plot.tree(ddr.tree.full)
ggsave("04_Expression/results/ddrtree_gene_expression.svg", width=6, height=6)

options(repr.plot.width=6, repr.plot.height=6)
plot.tree(ddr.tree.full, srs.info$SRSq, "SRSq")
ggsave("04_Expression/results/ddrtree_gene_expression_srsq.svg", width=6, height=6)
```

The tree for the eigengene expression matrix. Each dot represents a branch point (a vertex with degree > 2). The black path represents the underlying graph.

```r
options(repr.plot.width=6, repr.plot.height=6)
plot.tree(ddr.tree.eigen)
ggsave("04_Expression/results/ddrtree_eigengene_expression.svg", width=6, height=6)

options(repr.plot.width=6, repr.plot.height=6)
plot.tree(ddr.tree.eigen, srs.info$SRSq, "SRSq")
ggsave("04_Expression/results/ddrtree_eigengene_expression_srsq.svg", width=6, height=6)
```

## Identify Patient Clusters on Tree

Since the tree has defined branches, we can separate the patients by branches and treat them as clusters. I will do this for the eigengene expression tree, which has multiple well-defined branches from visual inspection.

```r
Z = t(ddr.tree.eigen$Z) %>%
  as.data.frame() %>%
  dplyr::select(Comp1=1, Comp2=2)

Y = t(ddr.tree.eigen$Y) %>%
  as.data.frame() %>%
  dplyr::select(Comp1=1, Comp2=2)

tree = as.data.frame(summary(ddr.tree.eigen$stree)) %>%
  dplyr::filter(i < j) %>%
  dplyr::mutate(StartComp1=Y$Comp1[i], StartComp2=Y$Comp2[i], EndComp1=Y$Comp1[j], EndComp2=Y$Comp2[j])

Y$Degree = table(c(tree$i, tree$j))[1:nrow(Y)]
```

I identify branches using an iterative pruning approach. The potential set of points that need to be assigned to a branch are all the points in the tree that have degree 2. A point with degree 1 is a leaf point, and a point with degree > 2 is a branch point. These points to be assigned are stored in `point.set`.

In each iteration (as long as `point.set` still has elements), I sample the first point from `point.set`. I add this to a queue and perform a breadth-first search along the tree until I reach either leaf or branch points. This set of points is then considered one branch and added to the `branches` list. Leaf points are also added to branches, but branch points are labelled `"N/A"`.

```r
point.set <- setdiff(1:nrow(Y), which(Y$Degree > 2))
branches <- list()

while (length(point.set) > 0) {
  
  branch = list()
  
  nucleus = point.set[1]
  
  q = queue()
  pushback(q, nucleus)
  
  while (length(q) > 0) {
    
    sample.point = pop(q)
    branch[[length(branch) + 1]] = sample.point
    
    attached.points = setdiff(c(as.matrix(subset(tree, i == sample.point | j == sample.point)[,1:2])), sample.point)
    
    for (i in 1:length(attached.points)) {
      if (Y$Degree[attached.points[i]] <= 2 && !(attached.points[i] %in% branch)) {
        pushback(q, attached.points[i])
      }
    }
  }
  
  branches[[length(branches) + 1]] <- unlist(branch)
  point.set <- setdiff(point.set, unlist(branch))
}

tree.branches <- lapply(
    1:length(branches), 
    function(x) data.frame(Sample=branches[[x]], Branch=x)
  ) %>% 
  do.call(rbind, .) %>%
  rbind(., data.frame(Sample=which(Y$Degree > 2), Branch="N/A")) %>%
  dplyr::arrange(Sample) %>%
  dplyr::mutate(Branch=factor(Branch, levels=c("N/A", 1:length(branches))))
```

We can plot the tree with branch cluster labels to visualize the final clusters.

```r
plot.data <- Y %>%
  dplyr::mutate(Branch=tree.branches$Branch) %>%
  dplyr::filter(Branch != "N/A") %>%
  dplyr::group_by(Branch) %>%
  dplyr::summarize(Mid.Comp1=(min(Comp1) + max(Comp1))/2, Mid.Comp2=(min(Comp2) + max(Comp2))/2)

options(repr.plot.width=6, repr.plot.height=6)
plot.tree(ddr.tree.eigen) +
  geom_label(data=plot.data, aes(x=Mid.Comp1, y=Mid.Comp2, label=Branch))
ggsave("04_Expression/results/ddrtree_eigengene_expression_branches.svg", width=6, height=6)
```

```r
options(repr.plot.width=10, repr.plot.height=8)
tree.branches %>%
    dplyr::group_by(Branch) %>%
    dplyr::summarize(N.Samples=n()) %>%
    dplyr::filter(Branch != "N/A") %>%
    ggplot() +
    geom_bar(aes(x=Branch, y=N.Samples), stat="identity") +
    ggplot_theme
```

```r
srs.info %>%
  dplyr::mutate(Branch=tree.branches$Branch) %>%
  dplyr::filter(Branch != "N/A") %>%
  dplyr::group_by(Branch) %>%
  dplyr::summarize(SRS1=sum(SRS=="SRS1")/n(), SRS2=sum(SRS=="SRS2")/n(), SRS3=sum(SRS=="SRS3")/n()) %>%
  tidyr::gather("SRS", "Frequency", -Branch) %>%
  ggplot(aes(x=Branch, y=Frequency, fill=SRS)) +
  geom_bar(stat="identity", color="black") +
  scale_fill_brewer(palette="Pastel2") +
  ggplot_theme

sample.info %>%
  dplyr::mutate(Branch=tree.branches$Branch) %>%
  dplyr::filter(Branch != "N/A") %>%
  dplyr::filter(!is.na(diagnosis)) %>%
  dplyr::group_by(Branch) %>%
  dplyr::summarize(CAP=sum(diagnosis==1)/n(), FP=sum(diagnosis==2)/n()) %>%
  tidyr::gather("Diagnosis", "Frequency", -Branch) %>%
  ggplot(aes(x=Branch, y=Frequency, fill=Diagnosis)) +
  geom_bar(stat="identity", color="black") +
  scale_fill_brewer(palette="Pastel2") +
  ggplot_theme

cell.prop %>%
  dplyr::mutate(Branch=tree.branches$Branch) %>%
  dplyr::filter(Branch != "N/A") %>%
  dplyr::filter(!is.na(Neutrophils), !is.na(Lymphocytes), !is.na(Monocytes)) %>%
  dplyr::group_by(Branch) %>%
  dplyr::summarize(Neutrophils=mean(Neutrophils), Lymphocytes=mean(Lymphocytes), Monocytes=mean(Monocytes)) %>%
  dplyr::select(Branch, Neutrophils, Lymphocytes, Monocytes) %>%
  tidyr::gather("Cell.Type", "Frequency", -Branch) %>%
  ggplot(aes(x=Branch, y=Frequency, fill=Cell.Type)) +
  geom_bar(stat="identity", color="black") +
  scale_fill_brewer(palette="Pastel2") +
  ggplot_theme

eigengenes %>%
  dplyr::mutate(Branch=tree.branches$Branch) %>%
  dplyr::filter(Branch != "N/A") %>%
  ggplot() +
  geom_boxplot(aes(x=Branch, y=ME_42)) +
  ggplot_theme
```
